<h2>DRAWING ON IMAGES</h2><br>


```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#random image
blank_image = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
blank_image.shape
```

```python
plt.imshow(blank_image)
```

```python
#after creating a blank image we want to draw rectangle on it. so we use the rectangle() method. it takes the image first, then 2 sets of 
# points . i.e. the four corners of the rectanlge as input. in pt1 we provide the first 2 points acording to the size of the image and the same for pt2.
# we can choose the color for rectangle in format as below and in any color using the 3 combinations and similarly we can provide thickness to the rectangle
# in pixels
blank_image = cv2.rectangle(blank_image,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_image)
```

```python
cv2.rectangle(blank_image,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness=12)
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image,center=(150,150),radius=40,color=(0,255,0),thickness=12)
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image,center=(400,400),radius=40,color=(0,255,0),thickness=-1)
plt.imshow(blank_image)
```

```python
cv2.line(img=blank_image,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_image)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img=blank_image,text='MAYANK',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_image)
```

We can use CallBacks to connect images to event functions with OpenCV. This allows to directly interact with images and videos. Mainly done in python scripts.

```python
# so in this lecture we make a circle on an image . for that we first create a black image by ourselves, we can take any image we want. in the shjowing image section below we
# are creating an image and opening it just like we do in detail.py file. it creates a black image first then using cv2 we create a seprate window called my_drawing.
# and the window takes black image as input for showig. and the cv2.waitKey() command is used to kill a window when Esc is pressed.
# In the function section below, we first create a namedWindow which is used as a placeholder for images as is desribed in its docstring. this command is used to link a function
# that will be used to write on images. this function will auto create a link between that function and created window. Then the setMouseCallback function is used to create
# a mouse event. This will first be used as a mouse event function that will perform events on images opened in a window and take the function that will descirbe what events will
# happen on the image. Then the function itself is used to tell what kind of event will occur on the image. It will compulsorily take the paramters that is defined below, because 
# the cv2 function itself needs them. So we select a simple left button click event and on that click a circle is drawn with any dimensions and coloring as we want.
# the left button click method is written in if functino because we want circle to be drawn only when an event of leftbutton click happens.

##############
###FUNCTION###
##############

def draw_circle(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

#################################
####SHOWING IMAGE WITH OPENCV####
#################################
img = np.zeros((512,512,3),dtype=np.int8)

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
##############
###FUNCTION###
##############

def draw_circle(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,255,0),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(255,0,0),-1)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

#################################
####SHOWING IMAGE WITH OPENCV####
#################################
img = np.zeros((512,512,3),dtype=np.int8)

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
