```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
dog = cv2.imread('D:/Computer vision learning/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
plt.imshow(dog) # color is different because opencv and matplotlib expect different orders of the RGB channels
```

```python
# image read by matplotlib for RGB is => RGB
# opencv for rgb is => BGR
# so we have to transform
# to convert an BGR image to RGB image (which is supported by matplotlib)  we have to use below function of cvtColor. It transforms images from one color channel format to another
# we supply it the image first and then the channel from which 2 which we qwant to transform . in below case as is clearly written we convert from BGR 2 RGB
# . because we opened the image using opencv it opened in BGR format and we want to work on image using matpot, so we convert from BGR to RGB
fix_dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_dog)
```
