```python
# in this we draw on a video while camera is running. for this cell we simply choose a poistion on the frame to display a rectanlge.

import cv2

cap = cv2.VideoCapture(0)

# we take in the dimensions of video that can be captured by our camera and later use this to get a random area in which the rectanlge will be drawn
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# we take the top left corner coordinates at the middle of the video. x and y axis at top left of rectangle are at half of frame width and height
# top levft corner
x = width // 2
y = height // 2

# now we determine the length and breadth of our rectangle and here we take it at 4th of frame size and we later add this to get the bottom right coordinates of our rectangle

# width and height of rectangle
w = width // 4
h = height // 4

#bottom right corner will be x+w and y+h

# double divide sign will return integer of division. single divide sign will return in float if it occurs

while True:
    ret, frame = cap.read()

    cv2.rectangle(frame, pt1=(x,y),pt2=(x+w,y+h),color=(0,0,255), thickness=4) # we know this function we simply provide the values we get from above and a rectanle will be shown 
     # rest all is same
    cv2.imshow('frame', frame) # always draw rectangle first then show the frame

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2


#callback function will draw rectangle
def draw_rectangle(event,x,y,flags,params):
    
    global p1,pt2,topleft_clicked,botright_clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        #reset rectangle , checks if rectangle already there
        if topleft_clicked == True and botright_clicked == True:
            
            topleft_clicked = False
            botright_clicked = False
            pt1 = (0,0)
            pt2 = (0,0)
        
        if topleft_clicked == False:
            pt1 = (x,y)
            topleft_clicked = True
        
        elif botright_clicked == False:
            pt2 = (x,y)
            botright_clicked = True

# global variables
pt1 = (0,0)
pt2 = (0,0)
topleft_clicked = False
botright_clicked = False

# connect to callbacks
cap = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rectangle)

while True:
    ret, frame = cap.read()

    #drawwing on frame based off global varibales
    if topleft_clicked == True:
        cv2.circle(frame, center=pt1, radius=5, color=(0,0,255), thickness=-1)

    if topleft_clicked and botright_clicked:
        cv2.rectangle(frame, pt1, pt2, (0,0,255), 2)

    cv2.imshow('Test', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
# Create a function based on a CV2 Event (Left button click)

# mouse callback function
def draw_rectangle(event,x,y,flags,param):

    global pt1,pt2,topLeft_clicked,botRight_clicked

    # get mouse click
    if event == cv2.EVENT_LBUTTONDOWN:

        if topLeft_clicked == True and botRight_clicked == True:
            topLeft_clicked = False
            botRight_clicked = False
            pt1 = (0,0)
            pt2 = (0,0)

        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True
            
        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True

        
# Haven't drawn anything yet!

pt1 = (0,0)
pt2 = (0,0)
topLeft_clicked = False
botRight_clicked = False

cap = cv2.VideoCapture(0) 

# Create a named window for connections
cv2.namedWindow('Test')

# Bind draw_rectangle function to mouse cliks
cv2.setMouseCallback('Test', draw_rectangle) 


while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if topLeft_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0,0,255), thickness=-1)
        
    #drawing rectangle
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)
        
        
    # Display the resulting frame
    cv2.imshow('Test', frame)

    # This command let's us quit with the "q" button on a keyboard.
    # Simply pressing X on the window won't work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
