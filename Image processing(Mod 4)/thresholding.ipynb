<h2>Image thresholding</h2><br> In some CV applications it is often necessary to convert color images to grayscale , since only edges and shapes end up important. Similarly, some applications only require a binary image showing general shapes. <BR><br> Thresholding is fundamentally a very simple method of segmenting an image into differnet parts. Thresholding will convert an image to consist of only 2 values, white or black. 

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('rainbow.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
ret, thres1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
ret
```

```python
thres1
```

```python
plt.imshow(thres1, cmap='gray') # binary image
```

```python
ret, thres1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thres1, cmap='gray')
```

```python
ret, thres1 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
```

```python
plt.imshow(thres1, cmap='gray')
```

```python
ret, thres1 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
```

```python
plt.imshow(thres1, cmap='gray')
```

```python
ret, thres1 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thres1, cmap='gray')
```

```python
# Real world image thresholding

img = cv2.imread('crossword.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# simple function to display an image and increase its size a bit... and shows it in grayscale
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 160, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

<h1>BELOW PART IS THE IMPORTANT PART AND IS USED IN REAL APPS</h1>

```python
# IMPORTANT !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
```

```python
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
```

```python
show_pic(blended)
```
