An image gradient is a directional change in intensity or color in an image. We explore basic Sobel-Feldman operators. It uses 2 3*3 kernels which are convolved with original image to calculate approximations of derivatived - one for horizontal changes and one for vertical. 

```python
import  cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('sudoku.jpg', 0)
```

```python
display_img(img)
```

```python
sobel_x = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
```

```python
display_img(sobel_x)
```

```python
sobel_y = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
display_img(sobel_y)
```

```python
laplacian = cv2.Laplacian(img, ddepth=cv2.CV_64F)
display_img(laplacian)
```

```python
blended = cv2.addWeighted(src1=sobel_x, alpha=0.5, src2=sobel_y, beta=0.5, gamma=0)
```

```python
display_img(blended)
```

```python
ret, th1 = cv2.threshold(img, thresh=100, maxval=255, type=cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
