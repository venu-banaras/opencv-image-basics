Morphological operators are sets of kernels that can achieve variety of effects, such as reducing noise . Certain operators are very good at reducing black points on a white background(and vice versa). Certain operators can also achieve an erosion and dilation effect that can add or erode from an existing image. This effect is most easoly seen on text data,.... 

```python
import  cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_image():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='ABCDE', fontFace=font, fontScale=5,org=(50,300),color=(255,255,255),thickness=20)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_image()
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
result = cv2.erode(img,kernel, iterations=4)
```

```python
display_img(result)
```

```python
img = load_image()
```

```python
noise = np.random.randint(low=0, high=2,size=(600,600))
```

```python
noise
```

```python
display_img(noise)
```

```python
noise *= 255
```

```python
noise_img = noise + img
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

```python
img = load_image()
```

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
display_img(black_noise)
```

```python
black_noise *= -255
```

```python
black_noise_img = img + black_noise
```

```python
black_noise_img[black_noise_img == -255] = 0
```

```python
black_noise_img.min()
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
```

```python
display_img(closing)
```
